```python
#statewise groundwater reserve
state_list = []
total_frnd_water_recharge = []
curr_gw_extr_list = []
future_available_GW_list = []
#Net Ground Water Availability for future use

for state, subset in df.groupby('Name of State'):
    #print(state, sum(subset['Net Ground Water Availability for future use']))
    state_list.append(state)
    total_frnd_water_recharge.append(sum(subset['Total Annual Ground Water Recharge']))
    curr_gw_extr_list.append(sum(subset['Total Current Annual Ground Water Extraction']))
    future_available_GW_list.append(sum(subset['Net Ground Water Availability for future use']))
dfnew = pd.DataFrame({"State":state_list, "GW_Recharge":total_frnd_water_recharge, "GW_Extraction": curr_gw_extr_list, "Future_GW_Available": future_available_GW_list})
```

## Machine Learning

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv('/content/Dynamic_2017_2_0 (2).csv')

# Drop unnecessary columns
df.drop(['S.no.', 'Name of State', 'Name of District'], axis=1, inplace=True)

# Handle missing values
df.fillna(0, inplace=True)  # Fill NaN values with 0, you may choose a different strategy based on your data

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)  # Convert the scaled NumPy array back to a DataFrame

# Split the dataset into features (X) and target variable (y)
X = df_scaled.drop("Net Ground Water Availability for future use", axis=1)
y = df_scaled["Net Ground Water Availability for future use"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score

# Create and train the Linear Regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Make predictions on the test set
linear_reg_preds = linear_reg.predict(X_test)

# Evaluate the performance of the Linear Regression model
mse = mean_squared_error(y_test, linear_reg_preds)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, linear_reg_preds)
r2 = r2_score(y_test, linear_reg_preds)
explained_var = explained_variance_score(y_test, linear_reg_preds)

# Print the metrics
print("Linear Regression Metrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")
print(f"Accuracy of the Model: {explained_var}")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, explained_variance_score

# Define a function to print multiple metrics
def print_regression_metrics(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    explained_var = explained_variance_score(y_true, y_pred)

    print(f"{model_name} Metrics:")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"R-squared (R2): {r2}")
    print(f"Accuracy of the Model: {explained_var}")
    print("\n")
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming X_train, X_test, y_train, y_test are your training and testing data
# You should replace these with your actual data

# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
linear_reg_preds = linear_reg.predict(X_test)
print_regression_metrics(y_test, linear_reg_preds, "Linear Regression")

# Random Forest
random_forest = RandomForestRegressor()
random_forest.fit(X_train, y_train)
rf_preds = random_forest.predict(X_test)
print_regression_metrics(y_test, rf_preds, "Random Forest")

# Support Vector Regressor
svr = SVR()
svr.fit(X_train, y_train)
svr_preds = svr.predict(X_test)
print_regression_metrics(y_test, svr_preds, "Support Vector Regressor")

# Gradient Boosting Regressor
gradient_boosting = GradientBoostingRegressor()
gradient_boosting.fit(X_train, y_train)
gb_preds = gradient_boosting.predict(X_test)
print_regression_metrics(y_test, gb_preds, "Gradient Boosting Regressor")


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Assuming X_train, X_test, y_train, y_test are your training and testing data
# You should replace these with your actual data

# Define a simple neural network model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=1, activation='linear'))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model (you should replace this with your actual training data)
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Now you can use the trained model to make predictions
nn_preds = model.predict(X_test)
print_regression_metrics(y_test, nn_preds, "Neural Network")

```

```python
import matplotlib.pyplot as plt
import numpy as np
models = ["Linear Regression", "Random Forest", "SVR", "Gradient Boosting", "Neural Network"]

# Calculate R2 scores for each model
r2_scores = [r2_score(y_test, linear_reg_preds),
             r2_score(y_test, rf_preds),
             r2_score(y_test, svr_preds),
             r2_score(y_test, gb_preds),
             r2_score(y_test, nn_preds)]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

bars = plt.bar(np.arange(len(models)), r2_scores, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.xticks(np.arange(len(models)), models)
plt.ylabel('Accuracy Score')
plt.title('Comparison of Regression Models based on Performance')

# Add text annotations on each bar
for bar, r2_score_value in zip(bars, r2_scores):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 0.01,
             f'{r2_score_value:.3f}', fontsize=9, ha='center', va='bottom')

plt.show()
```
